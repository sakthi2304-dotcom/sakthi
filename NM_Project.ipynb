```python
import pandas as pd
df = pd.read_csv('food delivery.csv')
df.head()
```

```python
df['Revenue'] = df['Quantity'] * df['Price']
print("Total Revenue:", df['Revenue'].sum())
```

```python
df['Hour'] = pd.to_datetime(df['Time']).dt.hour

peak_orders = df.groupby('Hour').size()

print(peak_orders)
```

```python
import matplotlib.pyplot as plt

peak_orders.plot(kind='bar')

plt.title("Orders by Hour")
plt.xlabel("Hour")
plt.ylabel("Number of Orders")

plt.show()
```

```python
location_orders = df.groupby('Location').size()

print(location_orders)
```

```python
location_orders.plot(kind='pie', autopct='%1.1f%%')

plt.title("Orders by Location")

plt.show()
```

```python
popular_items = df.groupby('Item')['Quantity'].sum()
popular_items.sort_values(ascending=False).head(10).plot(kind='bar')

plt.title("Top 5 Food Items")
plt.xlabel("Food Items")
plt.ylabel("Quantity Sold")

plt.show()
```

```python
location_orders = df.groupby('Location').size()

print(location_orders)
```

```python
location_orders.plot(kind='pie', autopct='%1.1f%%')

plt.title("Orders by Location")

plt.show()
```

```python
plt.figure(figsize=(12,8))

plt.subplot(2,2,1)
peak_orders.plot(kind='bar')
plt.title("Peak Order Hours")

plt.subplot(2,2,2)
popular_items.sort_values(ascending=False).head(5).plot(kind='bar')
plt.title("Top Food Items")

plt.subplot(2,2,3)
location_orders.plot(kind='pie', autopct='%1.1f%%')
plt.title("Location Analysis")

plt.tight_layout()
plt.show()
```
